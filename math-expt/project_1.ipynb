```matlab
P = load('math-expt/data/p1_swanPolygon.txt');
x = P(:,1);
y = P(:,2);
```

```matlab
% add constraints
C=[];
for i=1:length(x)-1
    C = [C; i i+1];
end
C = [C; length(x) 1];
```

```matlab
DT = delaunayTriangulation(x,y);
DT.Constraints = C;
IO = isInterior(DT);
```

```matlab
DT(IO,:)
triplot(DT(IO,:),x,y,'r');
```
