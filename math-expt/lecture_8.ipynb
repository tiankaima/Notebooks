#### Task 1

```matlab
x = linspace(-2,2,20);
y = x';
z = abs(x.*y);
h = surf(x,y,z);
set(h, 'FaceColor',[1 0 0], 'FaceAlpha', 0.2, 'EdgeAlpha', 0.1)
hold on;

t = -2:0.01:2;
plot3(t,t,t.^2,'b');
plot3(t,-t,t.^2,'b');
plot3(t,zeros(length(t)),zeros(length(t)),'r');
plot3(zeros(length(t)),t,zeros(length(t)),'r');
axis equal;
hold off;
```

#### Task 2

X-Y Plane:

```matlab
for t=1:10
    ezplot(@(x,y) abs(x.*y)-t,[-5,5,-5,5]);
    hold on;
end
axis equal;
hold off;
```

Intersecting this surface with a plane parallel to the xy plane gives us the parabola.

Y-Z Plane and X-Z Plane share the same equation, thus the same graph:

```matlab
for t=1:10
    ezplot(@(x,y) y-abs(x*t),[-5,5,-5,5]);
    hold on;
end
axis equal;
hold off;
```

The surface is a ruled surface, for every point $(x_0,y_0,z_0)$ on the surface, consider line $l: z = z_0 / x_0 * x, y=y_0$, it's easy to shown that $l$ is on the surface and $l$ passes $(x_0,y_0,z_0)$.
