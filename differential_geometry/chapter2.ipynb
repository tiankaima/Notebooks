```
(* Global functions *)
CrossProduct2D[v1_List, v2_List] := v1[[1]]*v2[[2]] - v1[[2]]*v2[[1]]
```

```
ClearAll["Global`*"]
r[t] = {a*Cos[t], b*Sin[t]}
r1[t] = D[r[t], t]
r2[t] = D[r1[t], t]

k[t] = CrossProduct2D[r1[t], r2[t]]/CrossProduct2D[r1[t], r2[t]]^2;
Simplify[k[t]]
```

```
ClearAll["Global`*"]
x[t] = a*Cos[t];
y[t] = b*Sin[t];

x1[t] = D[x[t], t];
y1[t] = D[y[t], t];
x2[t] = D[x1[t], t];
y2[t] = D[y1[t], t];

k[t] = (x1[t]*y2[t] - y1[t]*x2[t])/(x1[t]^2 + y1[t]^2)^(3/2);

Simplify[k[t]]
```

```
ClearAll["Global`*"]
x[t] = a*Cosh[t];
y[t] = b*Sinh[t];

x1[t] = D[x[t], t];
y1[t] = D[y[t], t];
x2[t] = D[x1[t], t];
y2[t] = D[y1[t], t];

k[t] = (x1[t]*y2[t] - y1[t]*x2[t])/(x1[t]^2 + y1[t]^2)^(3/2);

Simplify[k[t]]
```
