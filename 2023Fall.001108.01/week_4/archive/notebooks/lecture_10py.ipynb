### Question

Draw Feigenbaum diagram of $x_{n+1}=rx_n(1-x_n)$, where $x\in (0,1]$, $r\in [0,4]$. What do you see?

### Solution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
@np.vectorize
def f(r, x):
    return r * x * (1 - x)

@np.vectorize
def s(r):
    x = np.random.rand()
    for i in range(100):
        x = f(r, x)
    return x

def draw_feigenbaum():
    r = np.linspace(0, 4, 10000)
    x = s(r)
    plt.scatter(r, x, s=0.1)

draw_feigenbaum()
```
